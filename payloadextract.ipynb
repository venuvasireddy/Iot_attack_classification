```python
import pandas as pd
import binascii
from scapy.all import rdpcap, TCP, Raw, IP

pcap_file ="d1_pcap/normal_train.pcap"
packets = rdpcap(pcap_file)

# List to store extracted payloads
mqtt_payloads = []

# Extract MQTT payloads
for pkt in packets:
    if pkt.haslayer(TCP) and pkt.haslayer(Raw) and pkt.haslayer(IP):
            raw_payload = pkt[Raw].load  # Extract raw payload
            payload_hex = binascii.hexlify(raw_payload).decode()
            mqtt_payloads.append({
                "Source IP": pkt[IP].src,
                "Destination IP": pkt[IP].dst,
                "Payload (Raw)": raw_payload,
                "Payload (Hex)": payload_hex
            })

df = pd.DataFrame(mqtt_payloads)
df_cleaned = df[~df["Payload (Hex)"].astype(str).isin(["c000", "d000","20020000","20020005","e000"])]

csv_filename = "WILL_DoS_1_filter.csv"
df_cleaned.to_csv(csv_filename, index=False)

print(f"Extracted {len(df)} MQTT payloads. Saved to '{csv_filename}'.")
```

```python
import pandas as pd
from collections import defaultdict

data = [
    "dataset4/BF1_DDoS_1.csv",
    "dataset4/BF1_DoS_1.csv",
    "dataset4/Delay_DDoS_1.csv",
    "dataset4/Delay_DoS_1.csv",
    "dataset4/Normal1.csv",
    "dataset4/Sub_DDoS_1.csv",
    "dataset4/Sub_DoS_1.csv",
    "dataset4/SYN_DDoS_1.csv",
    "dataset4/SYN_DoS_1.csv",
    "dataset4/WILL_DDoS_1.csv",
    "dataset4/WILL_DoS_1.csv"
]

feature_column = "Payload (Hex)"
payload_counts = defaultdict(int)  # Track occurrence in datasets

# Read each dataset and count unique payloads
for file in data:
    df = pd.read_csv(file)
    if feature_column in df.columns:
        unique_payloads = set(df[feature_column].dropna())
        for payload in unique_payloads:
            payload_counts[payload] += 1  # Track how many datasets contain this payload

a=[]
print("\nPayloads appearing in exactly two datasets:")
i=0
for payload, count in payload_counts.items():
    if count >= 2:
        a.append(payload)
        print(payload)
        i=i+1


print("Number of payloads appearing in exactly two datasets:", i)
```
